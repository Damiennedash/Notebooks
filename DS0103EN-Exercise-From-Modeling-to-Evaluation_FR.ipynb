Pour votre commodité, nous avons déjà placé les données sur un serveur IBM. Téléchargez les données depuis le serveur et lisez-les dans un DataFrame nommé **recipes**.


````python
# Téléchargement et chargement du fichier CSV contenant les recettes depuis le serveur cloud d'IBM
recipes = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0103EN-SkillsNetwork/labs/Module%202/recipes.csv")

print("Données chargées dans le DataFrame avec succès !") # Cela prend environ 30 secondes
````

Ensuite, répétez les étapes de prétraitement que vous avez implémentées dans le lab précédent **"De la Compréhension à la Préparation des Données"** afin de préparer les données pour la modélisation. Pour plus de détails sur cette préparation, veuillez vous référer à ce précédent lab.


````python
# 1. Correction du nom de la première colonne (qui contient le type de cuisine)
column_names = recipes.columns.values
column_names[0] = "cuisine"
recipes.columns = column_names

# 2. Conversion de tous les noms de cuisine en minuscules pour la cohérence
recipes["cuisine"] = recipes["cuisine"].str.lower()

# 3. Standardisation et regroupement des noms de cuisine (ex. regrouper 'irish' et 'uk-and-ireland' sous 'uk-and-irish')
recipes.loc[recipes["cuisine"] == "austria", "cuisine"] = "austrian"
recipes.loc[recipes["cuisine"] == "belgium", "cuisine"] = "belgian"
recipes.loc[recipes["cuisine"] == "china", "cuisine"] = "chinese"
recipes.loc[recipes["cuisine"] == "canada", "cuisine"] = "canadian"
recipes.loc[recipes["cuisine"] == "netherlands", "cuisine"] = "dutch"
recipes.loc[recipes["cuisine"] == "france", "cuisine"] = "french"
recipes.loc[recipes["cuisine"] == "germany", "cuisine"] = "german"
recipes.loc[recipes["cuisine"] == "india", "cuisine"] = "indian"
recipes.loc[recipes["cuisine"] == "indonesia", "cuisine"] = "indonesian"
recipes.loc[recipes["cuisine"] == "iran", "cuisine"] = "iranian"
recipes.loc[recipes["cuisine"] == "italy", "cuisine"] = "italian"
recipes.loc[recipes["cuisine"] == "japan", "cuisine"] = "japanese"
recipes.loc[recipes["cuisine"] == "israel", "cuisine"] = "jewish"
recipes.loc[recipes["cuisine"] == "korea", "cuisine"] = "korean"
recipes.loc[recipes["cuisine"] == "lebanon", "cuisine"] = "lebanese"
recipes.loc[recipes["cuisine"] == "malaysia", "cuisine"] = "malaysian"
recipes.loc[recipes["cuisine"] == "mexico", "cuisine"] = "mexican"
recipes.loc[recipes["cuisine"] == "pakistan", "cuisine"] = "pakistani"
recipes.loc[recipes["cuisine"] == "philippines", "cuisine"] = "philippine"
recipes.loc[recipes["cuisine"] == "scandinavia", "cuisine"] = "scandinavian"
recipes.loc[recipes["cuisine"] == "spain", "cuisine"] = "spanish_portuguese"
recipes.loc[recipes["cuisine"] == "portugal", "cuisine"] = "spanish_portuguese"
recipes.loc[recipes["cuisine"] == "switzerland", "cuisine"] = "swiss"
recipes.loc[recipes["cuisine"] == "thailand", "cuisine"] = "thai"
recipes.loc[recipes["cuisine"] == "turkey", "cuisine"] = "turkish"
recipes.loc[recipes["cuisine"] == "vietnam", "cuisine"] = "vietnamese"
recipes.loc[recipes["cuisine"] == "uk-and-ireland", "cuisine"] = "uk-and-irish"
recipes.loc[recipes["cuisine"] == "irish", "cuisine"] = "uk-and-irish"


# 4. Suppression des pays/cuisines qui ont moins de 50 recettes au total pour éviter le bruit
recipes_counts = recipes["cuisine"].value_counts()
cuisines_indices = recipes_counts > 50

cuisines_to_keep = list(np.array(recipes_counts.index.values)[np.array(cuisines_indices)])
recipes = recipes.loc[recipes["cuisine"].isin(cuisines_to_keep)]

# 5. Conversion des variables de présence d'ingrédients : "Yes" devient 1, "No" devient 0
recipes = recipes.replace(to_replace="Yes", value=1)
recipes = recipes.replace(to_replace="No", value=0)
````

## [bamboo_tree] Uniquement les cuisines asiatiques et indiennes

Ici, nous allons créer un arbre de décision uniquement pour un sous-ensemble de cuisines asiatiques (coréenne, japonaise, chinoise, thaïlandaise) et indienne. En effet, un arbre de décision ne fonctionne pas de manière optimale lorsque les données sont fortement déséquilibrées (biaisées) en faveur d'une seule cuisine ou d'un seul groupe de cuisines (comme c'est le cas ici avec la cuisine américaine qui domine le dataset). Nous pouvons soit exclure la cuisine américaine de l'analyse, soit construire des arbres de décision sur des sous-ensembles spécifiques de données. Nous choisisons la deuxième solution.


Construisons notre arbre de décision en utilisant les données relatives aux cuisines asiatiques et indiennes, et nommons cet arbre de décision *bamboo_tree*.


````python
# 1. Sélection du sous-ensemble de cuisines asiatiques et indiennes
asian_indian_recipes = recipes[recipes.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]

# 2. Séparation de la colonne cible (cuisine) et des caractéristiques d'entrée (ingrédients)
cuisines = asian_indian_recipes["cuisine"]
ingredients = asian_indian_recipes.iloc[:,1:] # Toutes les colonnes sauf la première (les ingrédients)

# 3. Initialisation du classificateur d'arbre de décision avec une profondeur maximale de 3
bamboo_tree = tree.DecisionTreeClassifier(max_depth=3)

# 4. Entraînement (ajustement) du modèle sur nos données
bamboo_tree.fit(ingredients, cuisines)

print("Modèle d'arbre de décision sauvegardé sous la variable bamboo_tree !")
````

Affichons graphiquement cet arbre de décision pour examiner sa structure et les critères de division qu'il a appris.


````python
# Exportation de l'arbre au format DOT pour la visualisation
export_graphviz(bamboo_tree,
                feature_names=list(ingredients.columns.values),
                out_file="bamboo_tree.dot",
                class_names=np.unique(cuisines),
                filled=True, # Remplir les nœuds de couleurs selon leur pureté/classe dominante
                node_ids=True, # Afficher les IDs des nœuds
                special_characters=True,
                impurity=False, # Masquer l'impureté de Gini pour simplifier l'affichage
                label="all",
                leaves_parallel=False)

# Lecture du fichier DOT et affichage graphique dans le notebook
with open("bamboo_tree.dot") as bamboo_tree_image:
    bamboo_tree_graph = bamboo_tree_image.read()
graphviz.Source(bamboo_tree_graph)
````

L'arbre de décision a appris les règles suivantes :
* Si une recette contient du **cumin** et du **poisson** (*fish*) mais **pas** de **yaourt** (*yoghurt*), il s'agit très probablement d'une recette **thaïlandaise** (*Thai*).
* Si une recette contient du **cumin** mais **pas** de **poisson** ni de **sauce soja** (*soy_sauce*), il s'agit très probablement d'une recette **indienne** (*Indian*).


Vous pouvez analyser les autres branches de l'arbre pour en déduire des règles similaires permettant d'identifier la cuisine des autres recettes. 


N'hésitez pas à sélectionner un autre sous-ensemble de cuisines et à construire un arbre de décision à partir de leurs recettes. Vous pouvez par exemple sélectionner des cuisines européennes et analyser les ingrédients clés qui permettent de les distinguer.


# Évaluation du Modèle <a id="4"></a>


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0103EN-SkillsNetwork/labs/Module%204/images/flowchart_evaluation.png" width="500">


Pour évaluer la performance de notre modèle sur les cuisines asiatiques et indiennes, nous allons diviser le jeu de données en un ensemble d'entraînement et un ensemble de test (méthode du *Train/Test Split*). Nous construirons l'arbre de décision à l'aide de l'ensemble d'entraînement. Ensuite, nous testerons le modèle sur l'ensemble de test et comparerons les cuisines prédites par le modèle avec les vraies origines des recettes. 


Créons d'abord un nouveau DataFrame contenant uniquement les données relatives aux cuisines asiatiques et indiennes. Appelons ce DataFrame **bamboo**.


````python
# Sélection des lignes correspondant uniquement aux 5 cuisines cibles
bamboo = recipes[recipes.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
````

Voyons maintenant combien de recettes sont disponibles pour chaque type de cuisine.


````python
# Comptage du nombre de recettes par classe pour observer la distribution
bamboo["cuisine"].value_counts()
````

Retirons 30 recettes de chaque cuisine pour les utiliser comme ensemble de test. Nommons cet ensemble de test **bamboo_test**.


````python
# Définition de la taille de l'échantillon de test par classe
sample_n = 30
````

Créez ensuite un DataFrame contenant 30 recettes sélectionnées aléatoirement pour chaque cuisine.


````python
# 1. Sélection aléatoire de 30 recettes de chaque cuisine pour constituer le jeu de test
random.seed(1234) # Fixer la graine aléatoire pour obtenir des résultats reproductibles
bamboo_test = bamboo.groupby("cuisine", group_keys=False).apply(lambda x: x.sample(sample_n))

# 2. Séparation des caractéristiques d'entrée (ingrédients du test) et de la variable cible (vraies cuisines)
bamboo_test_ingredients = bamboo_test.iloc[:,1:] # Les ingrédients du jeu de test
bamboo_test_cuisines = bamboo_test["cuisine"] # Les étiquettes (labels) correspondantes
````

Vérifiez qu'il y a bien exactement 30 recettes pour chaque cuisine dans l'ensemble de test.


````python
# Vérification du nombre d'échantillons de test par classe
bamboo_test["cuisine"].value_counts()
````

Créez à présent l'ensemble d'entraînement en retirant les lignes du jeu de test du DataFrame **bamboo**. Nommons cet ensemble d'entraînement **bamboo_train**.


````python
# 1. Identification des index des lignes utilisées pour le jeu de test
bamboo_test_index = bamboo.index.isin(bamboo_test.index)

# 2. Exclusion des données de test pour obtenir les données d'entraînement (~ signifie NON/inverse)
bamboo_train = bamboo[~bamboo_test_index]

# 3. Séparation des variables prédictives et cibles pour le jeu d'entraînement
bamboo_train_ingredients = bamboo_train.iloc[:,1:] # Ingrédients d'entraînement
bamboo_train_cuisines = bamboo_train["cuisine"] # Cuisines associées pour l'apprentissage
````

Vérifiez qu'il y a bien exactement 30 recettes de moins pour chaque cuisine dans ce jeu d'entraînement.


````python
# Vérification des nouveaux effectifs de notre jeu d'entraînement
bamboo_train["cuisine"].value_counts()
````

Construisons l'arbre de décision en utilisant le jeu d'entraînement **bamboo_train**, et appelons l'arbre généré **bamboo_train_tree**.


````python
# 1. Création d'un arbre de décision plus profond (max_depth=15) pour capturer des règles plus détaillées
bamboo_train_tree = tree.DecisionTreeClassifier(max_depth=15)

# 2. Entraînement du modèle uniquement sur le jeu d'entraînement
bamboo_train_tree.fit(bamboo_train_ingredients, bamboo_train_cuisines)

print("Le modèle d'arbre de décision entraîné a été sauvegardé sous la variable bamboo_train_tree !")
````

Affichons ce nouvel arbre de décision complexe pour l'explorer.


````python
# Exportation et génération visuelle de l'arbre plus complexe de profondeur 15
export_graphviz(bamboo_train_tree,
                feature_names=list(bamboo_train_ingredients.columns.values),
                out_file="bamboo_train_tree.dot",
                class_names=np.unique(bamboo_train_cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("bamboo_train_tree.dot") as bamboo_train_tree_image:
    bamboo_train_tree_graph = bamboo_train_tree_image.read()
graphviz.Source(bamboo_train_tree_graph)
````

Maintenant que vous avez configuré l'arbre pour être plus profond, beaucoup plus de nœuds de décision ont été générés.


#### Testons maintenant les performances de notre modèle sur les données de test.


````python
# Utilisation de l'arbre entraîné pour prédire les cuisines des ingrédients du jeu de test
bamboo_pred_cuisines = bamboo_train_tree.predict(bamboo_test_ingredients)
````

Pour quantifier la capacité de notre arbre de décision à classer correctement chaque recette, nous allons créer une **matrice de confusion**. Celle-ci présente un résumé synthétique du nombre de recettes correctement classées par origine. Elle met également en évidence quelles cuisines ont tendance à être confondues avec d'autres.


Créons donc la matrice de confusion pour évaluer la précision de la classification de nos recettes de test (`bamboo_test`).


````python
# 1. Récupération des étiquettes uniques de nos cuisines de test
test_cuisines = np.unique(bamboo_test_cuisines)

# 2. Calcul de la matrice de confusion brute
bamboo_confusion_matrix = confusion_matrix(bamboo_test_cuisines, bamboo_pred_cuisines, labels = test_cuisines)
title = 'Matrice de Confusion Bamboo'
cmap = plt.cm.Blues

# 3. Normalisation de la matrice en pourcentages (chaque ligne somme à 100%)
plt.figure(figsize=(8, 6))
bamboo_confusion_matrix = (
    bamboo_confusion_matrix.astype('float') / bamboo_confusion_matrix.sum(axis=1)[:, np.newaxis]
    ) * 100

# 4. Affichage de la matrice sous forme de carte de chaleur colorée
plt.imshow(bamboo_confusion_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar()
tick_marks = np.arange(len(test_cuisines))
plt.xticks(tick_marks, test_cuisines)
plt.yticks(tick_marks, test_cuisines)

# 5. Remplissage des cellules avec les valeurs numériques correspondantes
fmt = '.2f'
thresh = bamboo_confusion_matrix.max() / 2.
for i, j in itertools.product(range(bamboo_confusion_matrix.shape[0]), range(bamboo_confusion_matrix.shape[1])):
    plt.text(j, i, format(bamboo_confusion_matrix[i, j], fmt),
             horizontalalignment="center",
             color="white" if bamboo_confusion_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('Vraie cuisine (True label)')
plt.xlabel('Cuisine prédite (Predicted label)')

plt.show()
````

Après avoir exécuté le code ci-dessus, vous devriez obtenir une matrice de confusion semblable à celle-ci :

Les lignes représentent les vraies origines des recettes du jeu de données (les classes réelles) et les colonnes représentent les prédictions faites par le modèle. La somme de chaque ligne est égale à 100%. À l'aide de cette matrice de confusion de référence, nous pouvons faire les observations suivantes :

* En analysant la première ligne (Chinois - **Chinese**) : 60 % des recettes chinoises de notre jeu de test ont été classées correctement par l'arbre de décision. Cependant, 37 % des recettes chinoises ont été classées à tort comme **Coréennes** (*Korean*) et 3 % ont été classées à tort comme **Indiennes** (*Indian*).

* En analysant la ligne de la cuisine **Indienne** (*Indian*) : 77 % de ces recettes ont été classées correctement, tandis que 3 % ont été confondues avec la cuisine chinoise, 13 % avec la cuisine coréenne et 7 % avec la cuisine thaïlandaise.
